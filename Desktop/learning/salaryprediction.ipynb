```python
#importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

```python
#loading the csv
df = pd.read_csv("salary_data.csv")
#defining the feature(x) and target(y)
df

X= df[["YearsExperience"]]
y = df["Salary"]

#split it to train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.20,random_state = 42)

#creating a model
model = LinearRegression()
model.fit(X_train,y_train)

#make predictions
my_expe = [[2]]
predictions = model.predict(my_expe)

print(f"prediction for {my_expe[0]} year(s) of experience is",predictions)
print("actual:",y_test[1])

#visualization
import matplotlib.pyplot as plt
plt.scatter(X,y)#plot actual data points
plt.plot(X,model.predict(X))
plt.title("Salary VS years of experience")
plt.xlabel("years")
plt.ylabel("salary")

plt.show()






```
